```python
import spacy
from gliner_spacy.pipeline import GlinerSpacy
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("gliner_spacy", config = {"labels": ["cdsid", "organization"]})
```

```python
dict = {
    'cdsid':[] , "organization":[]
}
```

```python
valid_labels = {"cdsid", "organization"}
```

```python

for doc in nlp.pipe(lines, batch_size=32):
    ents = doc.ents  # because style="ent"
    for ent in ents:

      label = ent.label_.lower()
      if label in dict:
          dict[label].append(ent.text.strip())
```
